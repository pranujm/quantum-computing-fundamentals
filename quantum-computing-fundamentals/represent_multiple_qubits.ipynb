```python
# import all necessary libraries
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create a quantum circuit with 2 qubits
qc = QuantumCircuit(2)
qc.h(0) # H-gate to put first qubit in superposition
```

```python
# get the quantum state
state = Statevector.from_instruction(qc)

# plot both qubits on bloch spheres
plot_bloch_multivector(state, title = "First qubit in Superposition")
plt.show()
```

```python
# create a quantum circuit with 2 qubits
qc2 = QuantumCircuit(2)
```

```python
# get the quantum state
state = Statevector.from_instruction(qc2)

# plot both qubits on bloch spheres
plot_bloch_multivector(state, title = "Both qubits in |0 state")
plt.show()
```

```python
# create a quantum circuit with 2 qubits
qc3 = QuantumCircuit(2)
qc3.x(0) #X-gate to put first qubit in |1 state
qc3.x(1) #X-gate to put second qubit in |1 state
```

```python
# get the quantum state
state = Statevector.from_instruction(qc3)

# plot both qubits on bloch spheres
plot_bloch_multivector(state, title = "Both qubits in |1 state")
plt.show()
```
