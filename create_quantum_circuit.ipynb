```python
# import all necessary libraries
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
```

```python
# create a quantum circuit with one qubit and one classical bit
qr = QuantumRegister(1)
cr = ClassicalRegister(1)
circuit = QuantumCircuit(qr, cr)
print(circuit)
```

```python
# measure the qubit and store the result in the classical bit
circuit.measure(qr[0], cr[0])
print(circuit)
```

```python
# simulate 1,000-plus shots and plot the results on a histogram
simulator = AerSimulator()
job = simulator.run(circuit, shots=1024)
result = job.result()
counts = result.get_counts()

# plot histogram
plot_histogram(counts)
```

```python
# display the state vector and bloch sphere representation
# create circuit without measurement for statevector analysis
statevector_circuit = QuantumCircuit(1)

# get statevector
statevector = Statevector.from_instruction(statevector_circuit)

# display state vector in LaTeX format
array_to_latex(statevector)

# plot bloch sphere
plot_bloch_multivector(statevector)
```
